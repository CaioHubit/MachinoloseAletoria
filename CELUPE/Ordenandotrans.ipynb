```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Preparando o Dataframe
df = pd.read_csv('DADOS CCNF.txt', delimiter='\t', decimal=',')
df = df.drop(columns=['#'])
#display(df.to_string())
#df.astype({"Transmitancia":"int32"})
df = df.sort_values(by='Transmitancia', ascending=False)

df
```

```python
df.to_csv('dados_ordenados.txt', index=False, sep='\t', decimal=',')

print("DataFrame salvo como 'dados_ordenados.txt'")
```

```python
# Criar um gráfico de barras para todas as amostras de transmitância
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.bar(df.index, df['Transmitancia'])
plt.xlabel('Name')
plt.ylabel('Transmitancia')
plt.title('Transmitância de Todas as Amostras')
plt.xticks(df.index, df['Name'], rotation=90)  # Rótulos do eixo X
plt.tight_layout()  # Para evitar corte de rótulos
plt.show()
```

```python
novo_dfsem = df[~df['Name'].str.contains('[BCD]')]

top_10 = novo_dfsem.sort_values(by='Transmitancia', ascending=False).head(100)

# Criar um gráfico de barras para as 10 maiores amostras de transmitância
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Transmitancia'])
plt.xlabel('Amostras')
plt.ylabel('Transmitancia')
plt.title('As 10 Maiores Amostras de Transmitância')
plt.xticks(rotation=45)  # Rotacionar os rótulos para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python
# Classificar o DataFrame em ordem decrescente pela coluna 'Transmitancia' e selecionar as 10 maiores amostras
top_10 = df.sort_values(by='Transmitancia', ascending=False).head(100)

# Criar um gráfico de barras para as 10 maiores amostras de transmitância
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Transmitancia'])
plt.xlabel('Amostras')
plt.ylabel('Transmitancia')
plt.title('As 10 Maiores Amostras de Transmitância')
plt.xticks(rotation=45)  # Rotacionar os rótulos para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python

# Criando um novo DataFrame com amostras que têm valores de Concentracao entre 2 e 10
df_novasA = df[(df['Concentracao'] >= 2) & (df['Concentracao'] <= 10) & df['Name'].str.contains('[BCD]')]

(df_novasA)
```

```python

# Criando um novo DataFrame excluindo as linhas com "B", "C" ou "D" em "Name"
novo_dfsem = df[~df['Name'].str.contains('[BCD]')]

# Visualizando o novo DataFrame
print(novo_dfsem)
```

```python
# Classificar o DataFrame em ordem decrescente pela coluna 'Transmitancia' e selecionar as 10 maiores amostras
top_10semBCD = novo_dfsem.sort_values(by='Transmitancia', ascending=True).head(100)

# Criar um gráfico de barras para as 10 maiores amostras de transmitância
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Transmitancia'])
plt.xlabel('Amostras')
plt.ylabel('Transmitancia')
plt.title('As 10 Maiores Amostras de Transmitância')
plt.xticks(rotation=45)  # Rotacionar os rótulos para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python

# Agrupe por 'Concentracao' e ordene 'Transmitancia' em ordem crescente dentro de cada grupo
novo_df_agrupado = novo_dfsem.sort_values(by='Transmitancia').groupby('Concentracao')

# Obtenha o primeiro valor de 'Transmitancia' em cada grupo
novo_df_agrupado = novo_df_agrupado.first().reset_index()

# Visualize o novo DataFrame agrupado e ordenado
print(novo_df_agrupado)
```

```python
top_10 = novo_df_agrupado.sort_values(by='Transmitancia', ascending=False).head(100)

# Criar um gráfico de barras para as 10 maiores amostras de transmitância
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Transmitancia'])
plt.xlabel('Amostras')
plt.ylabel('Transmitancia')
plt.title('As 10 Maiores Amostras de Transmitância')
plt.xticks(rotation=45)  # Rotacionar os rótulos para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python

# Filtrar o DataFrame para conter apenas os dados com Concentracao igual a 2
novo_df_concentracao_2 = novo_df_agrupado[novo_df_agrupado['Concentracao'] == 2]

# Visualizar o novo DataFrame
print(novo_df_concentracao_2)
```

```python
top_10 = novo_df_concentracao_2.sort_values(by='Transmitancia', ascending=False).head(10)

# Criar um gráfico de barras para as 10 maiores amostras de transmitância
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Transmitancia'])
plt.xlabel('Amostras')
plt.ylabel('Transmitancia')
plt.title('As 10 Maiores Amostras de Transmitância')
plt.xticks(rotation=45)  # Rotacionar os rótulos para facilitar a leitura
plt.tight_layout()
plt.show()
```
